#### 3. Invariant Metrics

An invariant metric is used in the sanity check during A/B Testing, which should be consistent before and after the experiment. 

In this project, since the experimental message would not show up until a student press the "Start free trial" button, any metric collected before the enrollment step would not be affected by the experiment. As a result, I chose  **number of cookies**, **number of clicks**, and **click-through probability** to be the invariant metrics.  

#### 4. Evaluation Metrics

An evaluation metric is used to measure the business impact of an experiment.

Looking at the definitions of the given metrics, the **Gross Conversion** rate would give us an idea on the percentage of students who eventually enrolled in the free trial after seeing the pop-up experimental message. 

The **Retention** rate would be a good evaluation metric as well, because it reflects the percentage of students who made at least one payment after the enrollment which would be affected by the experimental message. 

Last but not least, the **Net Conversion** rate should be included, as it shows us the percentage of students who made at least one payment out of the total number of cookies to click the "Start free trial" where students would see the experimental message. 

### Measuring Standard Deviation

For each evaluation metric, we were asked to calculate the **analytic estimate** standard deviation, given a sample of 5000 cookies visiting the course overview page and the [baseline values](https://docs.google.com/spreadsheets/d/1MYNUtC47Pg8hdoCjOXaHqF-thheGpUshrFA21BAJnNc/edit#gid=0). It should be noted that we were actually measuring the standard error of sample mean. 

```python
# Number of Cookies
baseline_cookies = baseline.values[0,1]

# Number of Free Trial Cookies
baseline_free_trial_cookies = baseline.values[1,1]

# Click-through Probability
CTP = baseline.values[3,1]

# Gross Conversion
baseline_gc = baseline.values[4,1]

# Retetion 
baseline_rt = baseline.values[5,1]

# Net Conversion
baseline_nc = baseline.values[6,1]

# Sample Size 
sample_size = 5000
```

Calculate the value of unit of analysis for each metric based on the given sample size: 

```python
# Number of free trial cookies
free_trial_cookies = sample_size*CTP
free_trial_cookies
```

```python
# Number of enrolled user-ids
enrolled_user_ids = free_trial_cookies*baseline_gc
enrolled_user_ids
```

Under the assumption that the data are independent and identically distributed or i.i.d., we can assume the data follow a _Bernoulli_ distribution. Based on _Central Limit Therom_, the formula used to calculate the standard error of the sample is shown below: 

$$ {s}_\bar{x} = \sqrt{\frac{\hat{p}(1-\hat{p})}{n}}$$

```python
# se^ for Gross Conversion 
se_hat_gc = np.sqrt(baseline_gc*(1-baseline_gc)*(1/free_trial_cookies))
se_hat_gc
```

```python
# se^ for Retention
se_hat_rt = np.sqrt(baseline_rt*(1-baseline_rt)*(1/enrolled_user_ids))
se_hat_rt
```

```python
# se^ for Net Conversion 
se_hat_nc = np.sqrt(baseline_nc*(1-baseline_nc)*(1/free_trial_cookies))
se_hat_nc
```

### Sizing

#### 1. Number of Samples vs. Power

Before we calculate the number of page views required in this experiment, we need to decided whether or not to use the _Bonferroni Correction_, which is used to counteract the problem of multiple comparisons. Since there are three metrics, given 95% confidence interval, the probability of getting at least one false positive result is:  

```python
1-(0.95)**3
```

Whiile 14.6% is not that bad, it is in fact an overestimate because the three metrics are not independent at all. Besides that, the Bonferroni Correction is a very conservative method. Thus, I decided not to pursue this approach. 

From this formula:

$$ d_{min} = z_{1-\frac{\alpha}{2}}\cdot \sqrt{\frac{p_1 (1-p_1)}{n}} + z_{1-\beta}\cdot \sqrt{\frac{p_2 (1-p_2)}{n}}$$

We can obtain the sample size formula: 

$$ n = (\frac{z_{1-\frac{\alpha}{2}}\cdot \sqrt{p_1 (1-p_1)} + z_{\beta}\cdot \sqrt{p_2 (1-p_2)}}{d_{min}})^2 $$

Reference: [A/B Testing, from scrath](http://www.alfredo.motta.name/ab-testing-from-scratch/#easy-footnote-bottom-22-982)

```python
def sample_size_calculator (alpha, beta, p, delta):
    # p is the baseline conversion rate
    # delta, or dmin, is the practical significance level 
    
    if p > 0.5:
        p2 = p - delta
    else: 
        p2 = p + delta
    
    t_alpha2 = stats.norm.isf(1-alpha/2)
    
    t_beta = stats.norm.isf(1-beta)

    sd1 = np.sqrt(p * (1 - p) + p  * (1 - p ))
    sd2 = np.sqrt(p * (1 - p) + p2 * (1 - p2))
    
    return int(round((t_alpha2 * sd1 + t_beta * sd2)**2 / delta**2, 0))
```

It should be noted that the sample size caclulator only produces the sample size of one group of the experiment, so we need to double it in order to get the size for both control and treatment groups. 

More importantly, the sample size calculator is based on the unit of analysis of each metric, which means we need to work restrospectively by applying the reciprocals of the rate(s) to the results we got from the sample size calculator. 

```python
# Number of Page View Required for Gross Conversion 
num_free_trial_per_group = sample_size_calculator(0.05, 0.2, baseline_gc, 0.01)
ss_gc = int(round(num_free_trial_per_group*2/CTP, 0))
ss_gc
```

```python
# Number of Page View Required for Retention 
num_enrolled_user_ids_per_group = sample_size_calculator(0.05, 0.2, baseline_rt, 0.01)
ss_rt = int(round(num_enrolled_user_ids_per_group*2/baseline_gc/CTP, 0))
ss_rt
```

```python
# Number of Page View Required for Net Conversion 
num_free_trial_per_group2 = sample_size_calculator(0.05, 0.2, baseline_nc, 0.0075)
ss_nc = int(round(num_free_trial_per_group2*2/CTP, 0))
ss_nc
```

At this point, I would choose **4741212** to be the total number of page views required, because it is the greatest number among the three metrics.

## Part III: Experiment Analysis

### Sanity Check

Before we run an A/B test, it is important to perform sanity checks to make sure the control and experiment groups are comparable, such as no filters were added to the control group when collecting the data, and the data capture was actually capturing the events we wanted to look for.

First of all, we should check the **population sizing metric** based on the unit of diversion. This is to make sure the experimental population and the control population is comparable. Second of all, the **invariant metrics** we chose in Part II sould stay consistent before and after the experiment. 


_The data collected can be found at_ [Final Projecct Results](https://docs.google.com/spreadsheets/d/1Mu5u9GrybDdska-ljPXyBjTpdZIUev_6i7t4LRDfXM8/edit#gid=0)

```python
# Import the data collected 
# Note that I combined the control and experiment data and denoted them using the **Group** variable for the sake of the analysis. 
# Group == 0: data from the control group; Group == 1: data from the experiment group

dat = pd.read_csv('final_project_results.csv')
dat.head()
```

To begin with, I would like to check the **robustness** of the invariant metrics by looking their distribution histogram. 

As shown above, the distribution for the **number of cookies** and the **number of page views** seem to be normal, we should be able to calculate the analytic standard error readily. Refer to [Checking Invariants Part 2](https://classroom.udacity.com/courses/ud257/lessons/4085798776/concepts/40713087720923). 

However, the distribution for **CTP** appears to have two modes. So I decided to perform a bootstrapping test to reconstruct the distribution of CTP based on resampled data, and calculate the empirical confidence interval.   

```python
# Number of pagesviews 
plt.hist(dat.Pageviews, bins=15, color='lightblue')
plt.title('Distribution of the Page Views', fontsize=20)
plt.ylabel('Frequency', fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('Number of Page Views', fontsize=20)
plt.xticks(fontsize=20, rotation='vertical')
plt.show()
```

```python
# Sanity Check for Page Views

# Observed: Ncont / Ntotal
phat_views = dat[dat["Group"] == 0].sum()["Pageviews"]/dat.sum()["Pageviews"]

# Calculate the expected CIs based on CLT and the expected center 0.5
# SE based on the expected center of CI 0.5
p_views = 0.5
se_views = np.sqrt(p_views*(1-p_views)/(dat.sum()["Pageviews"]))

# CI
CI_lower_views = p_views - se_views*1.96
CI_upper_views = p_views + se_views*1.96

print(CI_lower_views, CI_upper_views, CI_lower_views < 0.5 and CI_upper_views > 0.5)
```

```python
# Number of cookies  
plt.hist(dat.Clicks, bins=15, color='skyblue')
plt.title('Distribution of the Unique Cookies', fontsize=20)
plt.ylabel('Frequency', fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('Number of Unique Cookies', fontsize=20)
plt.xticks(fontsize=20, rotation='vertical')
plt.show()
```

```python
# Sanity Check for Clicks

# Ncont / Ntotal
phat_clicks = dat[dat["Group"] == 0].sum()["Clicks"]/dat.sum()["Clicks"]

# SE based on the expected center of CI 0.5
se_clicks = np.sqrt(0.5*0.5/(dat.sum()["Clicks"]))

# CI
CI_lower_clicks = 0.5 - se_clicks*1.96
CI_upper_clicks = 0.5 + se_clicks*1.96

print(CI_lower_clicks, CI_upper_clicks, CI_lower_clicks < 0.5 and CI_upper_clicks > 0.5)
```

In the sanity check tests above, the two observed confidence intervals both contained 0.5, which means there were no significant differences in the invariant metrics before and after the experiment.

#### Sanity check for CTP

Now we can construct _bootstrapping_ test to check if the observed confidence intervals appeared by chance. 

```python
# CTP 
plt.hist(dat.Clicks/dat.Pageviews, bins=15, color='orange')
plt.title('Distribution of the Click-through Probability', fontsize=20)
plt.ylabel('Frequency', fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('Click-through Probability', fontsize=20)
plt.xticks(fontsize=20, rotation='vertical')
plt.show()
```

```python
# Boostrap relicate function 

def bootstrap_relicate_1d(data, func):
    # Generate bootstrap replicate of 1 dimensional data
    bs_sample = np.random.choice(data, len(data)) # bootstrap relicate, size = n
    return func(bs_sample)
```

```python
cont_ctp = dat[dat["Group"] == 0]["Clicks"]/dat[dat["Group"]==0]["Pageviews"]
exp_ctp  = dat[dat["Group"] == 1]["Clicks"]/dat[dat["Group"]==1]["Pageviews"]

v = np.vstack((cont_ctp, exp_ctp)) # create an array 
d = v[1] - v[0]

# Bootstrap replicates 
bs_replicates = np.empty(1000)

for i in range(1000):
    bs_replicates[i] = bootstrap_relicate_1d(d, np.mean)

conf_int = np.percentile(bs_replicates, [2.5, 97.5])

bci_lower = conf_int[0]
bci_upper = conf_int[1]

print(bci_lower, bci_upper)
```

```python
# Plot a histogram of bootstrap replicates

plt.hist(bs_replicates, bins=15, normed = True, color='lightblue')
plt.xlabel('Difference in CTP')
plt.ylabel('PDF')
plt.xticks(rotation='vertical')
plt.axvline(x=bci_lower, c='b')
plt.axvline(x=bci_upper, c='b')
plt.axvline(x=0, c='r')

plt.show()
```

As shown on the graph above, if we repeat the experiment over and over again, then the data for Click-through Probability (CTP) seem to be normally distributed. Since 0 is captured in the 95% confidence interval from the bootstrapping test, we can conclude that there is no significant difference between the control and experiment groups with respect to this invariant metric. 

In summary, all three invariant metrics passed the sanity check. It is safe for us to continue our analysis. 

### Result Analysis 

#### 1. Effect Size Tests

```python
# CTP 
plt.hist(df.Enrollments/df.Clicks, bins=15, color='skyblue')
plt.title('Distribution of the Click-through Rate', fontsize=20)
plt.ylabel('Frequency', fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('Click-through Probability', fontsize=20)
plt.xticks(fontsize=20, rotation='vertical')
plt.show()
```

```python
def check_sig (x1, x2, n1, n2, alpha, dmin):
    
    p_pool = (x1 + x2)/(n1 + n2)
    se_pool = np.sqrt(p_pool*(1-p_pool)*(1/n1 + 1/n2))
    t_alpha2 = stats.norm.isf(alpha/2)
    d = x2/n2 - x1/n1
    
    CI_lower = round(d - se_pool*t_alpha2, 4)
    CI_upper = round(d + se_pool*t_alpha2, 4)
    
    if CI_lower < 0 and CI_upper > 0:
        
        statistical_sig = 'N'
        practical_sig = 'N'   

    else: 
        statistical_sig = 'Y'
                
        if CI_lower < dmin and CI_upper > dmin:            
            practical_sig = 'N'   
            
        else: 
            practical_sig = 'Y'   
    
    print ('CI lower bound: ', CI_lower, '\nCI upper bound: ', CI_upper,
           '\nStatistical significance: ', statistical_sig, 
           '\nPractical significance: ', practical_sig)
```

```python
# Drop null values
df = dat.dropna()

# Effect Size Test for Gross Conversion 

Xcont_enroll = df[df["Group"] == 0].sum()["Enrollments"]
Xexp_enroll  = df[df["Group"] == 1].sum()["Enrollments"]

Ncont_clicks = df[df["Group"] == 0].sum()["Clicks"]
Nexp_clicks  = df[df["Group"] == 1].sum()["Clicks"]

check_sig(Xcont_enroll, Xexp_enroll, Ncont_clicks, Nexp_clicks, 0.05, 0.01)
```

```python
cont_gc = Xcont_enroll/Ncont_clicks
exp_gc  = Xexp_enroll/Nexp_clicks

v = np.vstack((cont_gc, exp_gc)) # create an array 
d = v[1] - v[0]

# Bootstrap replicates 
bs_replicates = np.empty(1000)

for i in range(1000):
    bs_replicates[i] = bootstrap_relicate_1d(d, np.mean)

conf_int = np.percentile(bs_replicates, [2.5, 97.5])

bci_lower = conf_int[0]
bci_upper = conf_int[1]

print(bci_lower, bci_upper)
```

```python
# Plot a histogram of bootstrap replicates

plt.hist(bs_replicates, bins=15, normed = True, color='lightblue')
plt.xlabel('Difference in CTP')
plt.ylabel('PDF')
plt.xticks(rotation='vertical')
plt.axvline(x=bci_lower, c='b')
plt.axvline(x=bci_upper, c='b')
plt.axvline(x=0, c='r')

plt.show()
```

```python
# Effect Size Test for Net Conversion 

Xcont_pay = df[df["Group"] == 0].sum()["Payments"]
Xexp_pay  = df[df["Group"] == 1].sum()["Payments"]

check_sig(Xcont_pay, Xexp_pay, Ncont_clicks, Nexp_clicks, 0.05, 0.01)
```

```python
# Plot the distribution of net conversion 
plt.hist(df.Payments/df.Clicks, bins=15, color='skyblue')
plt.title('Distribution of the ', fontsize=20)
plt.ylabel('Frequency', fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('Net Conversion', fontsize=20)
plt.xticks(fontsize=20, rotation='vertical')
plt.show()
```

```python
cont_nc = Xcont_pay/Ncont_clicks
exp_nc  = Xexp_pay/Nexp_clicks

v = np.vstack((cont_nc, exp_nc)) # create an array 
d = v[1] - v[0]

# Bootstrap replicates 
bs_replicates = np.empty(1000)

for i in range(1000):
    bs_replicates[i] = bootstrap_relicate_1d(d, np.mean)

conf_int = np.percentile(bs_replicates, [2.5, 97.5])

bci_lower = conf_int[0]
bci_upper = conf_int[1]

print(bci_lower, bci_upper)
```

```python
# Plot a histogram of bootstrap replicates

plt.hist(bs_replicates, bins=15, normed = True, color='lightblue')
plt.xlabel('Difference in CTP')
plt.ylabel('PDF')
plt.xticks(rotation='vertical')
plt.axvline(x=bci_lower, c='b')
plt.axvline(x=bci_upper, c='b')
plt.axvline(x=0, c='r')

plt.show()
```
