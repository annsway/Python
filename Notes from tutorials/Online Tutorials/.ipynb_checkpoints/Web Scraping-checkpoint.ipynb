# Parsing Example

**Goal**: re-create a YouTube link to an embedded YouTube video on a website. 

First of all, get the website link. 

```python
vid_src = article.find('iframe', class_ ='youtube-player')
print(vid_src) 
```

Find out the YouTube link by searching the 'src', similar to dictionary.

Unlike the above examples, we don't want to grab the 'text' or 'string' within this class, but we want to grab **the value of that source atrribute from that tab**, which acts like a dictionary. 

```python
vid_src = article.find('iframe', class_ ='youtube-player')['src']
print(vid_src) 
```

Second, get the YouTube ID. 

```python
vid_id = vid_src.split('/')
print(vid_id)
```

```python
vid_id = vid_src.split('/')[4]
print(vid_id)
```

```python
vid_id = vid_id.split('?')[0]
print(vid_id)
```

Finally, re-construct the YouTube link.

```python
yt_link = 'https://youtube.com/watch?v={}'.format(vid_id)
print(yt_link)
```

Print out all the links on the website

```python
for article in soup.find_all('article'):
    headline = article.h2.a.text
    print(headline)
    
    summary = article.find('div', class_='entry-content').p.text
    print(summary)
    
    try:
        vid_src = article.find('iframe', class_ = 'youtube-player')['src']

        vid_id = vid_src.split('/')[4]
        vid_id = vid_id.split('?')[0]

        yt_link = 'https://youtube.com/watch?v={}'.format(vid_id)

    except Exception as e:
        yt_link = None
        
    print(yt_link)
    
    print(' ')
```
